```python
def outer_function():
    x = 10  # Variable in the outer function's scope

    def inner_function():
        nonlocal x  # Declare x as nonlocal
        x = 20      # Modify the x from outer_function's scope
        print(f"Inner function: x = {x}")

    inner_function()
    print(f"Outer function: x = {x}")

outer_function()
```

```python
def evaluate_eligibility(age, has_degree, experience_years, project_score, is_manager):
  """
      For "and" Operator:
      print(True and True) # Output: True
      print(True and False) # Output: False
      print(False and True) # Output: False
      print(False and False) # Output: False

      For "or" operator:
      print(True or True) # Output: True
      print(True or False) # Output: True
      print(False or True) # Output: True
      print(False or False) # Output: False

      Args:

      Input:
           age: Age of candidate (int type)
           has_degree: Candidate degree if he/she has one (boolean type)
           experience_years: No. of years experience for candidate (int type)
           project_score: Project Score given by ATS system (float type)
           is_manager: Whether the candidate is a manager or not (boolean type)

      Output:
           overall_eligibility: If the candidate is eligible or not (boolean type)

  """

  # Condition 1: Basic requirements
  basic_requirements_met = (age >= 25 and has_degree)

  # Condition 2: Experience or high project score
  experience_or_score = (experience_years >= 5 or project_score > 80)

  # Condition 3: Managerial role with specific age or experience
  manager_specific_criteria = is_manager and (age > 30 or experience_years >= 10)

  # Combine all conditions for overall eligibility
  # A candidate is eligible if they meet basic requirements AND (experience/score OR manager criteria)
  overall_eligibility = basic_requirements_met and (experience_or_score or manager_specific_criteria)

  return overall_eligibility

# Test cases
print(f"Candidate 1 (Eligible): {evaluate_eligibility(28, True, 6, 75, False)}") # True: basic & experience
print(f"Candidate 2 (Eligible): {evaluate_eligibility(26, True, 3, 90, False)}") # True: basic & score
print(f"Candidate 3 (Eligible): {evaluate_eligibility(35, True, 8, 70, True)}") # True: basic & manager criteria (age > 30)
print(f"Candidate 4 (Eligible): {evaluate_eligibility(29, True, 12, 60, True)}") # True: basic & manager criteria (experience >= 10)

print(f"Candidate 5 (Not Eligible): {evaluate_eligibility(22, True, 5, 85, False)}") # False: age < 25 (fails basic)
print(f"Candidate 6 (Not Eligible): {evaluate_eligibility(30, False, 7, 70, False)}") # False: no degree (fails basic)
print(f"Candidate 7 (Not Eligible): {evaluate_eligibility(27, True, 2, 60, False)}") # False: fails experience/score and not manager
print(f"Candidate 8 (Not Eligible): {evaluate_eligibility(29, True, 4, 70, True)}")  # False: manager but fails specific age/experience
```

Exception Handling

Exception handling in Python provides a structured way to deal with runtime errors, or "exceptions," that can disrupt the normal flow of a program. Instead of crashing, a program can gracefully handle these unexpected events, allowing for error logging, user feedback, or corrective actions.

```python
try:
  # Code that might raise an exception
  result = 10 / 0
except ZeroDivisionError:
  # Code to handle ZeroDivisionError
  print("Cannot divide by zero!")
```

```python
try:
  value = int("abc")
except ValueError:
  print("Invalid input: cannot convert to integer.")
```

```python
try:


  num = int(input("Enter a number: "))
except ValueError:
  print("That's not a valid number!")
else:
  print(f"You entered: {num}")
```

```python
try:
  file = open("my_file.txt", "r")
  content = file.read()
except FileNotFoundError:
  print("File not found! No such file or directory. Please check the filename or directory name.")
finally:
  if 'file' in locals() and not file.closed:
    print(f"Content in the file: {content}")
    file.close()
    print("File closed.")
```

```python
import pandas as pd
import numpy as np
```

```python
try:
    df = pd.read_csv("https://raw.githubusercontent.com/Muralimekala/python/master/Salaries.csv")
except Exception as e:
    print(f"Error loading data: {e}")
    # Exit if data cannot be loaded
    exit()
```

```python
print("--- Data Loading and Cleaning Complete ---")
print(f"Data has {len(df)} rows and {len(df.columns)} columns.")
```

```python
print("\n" + "="*80)
print("1. Min, Average, Max BasePay Per Job Title")
print("="*80)
```

```python
basepay_summary = df.groupby('JobTitle')['BasePay'].agg(
    Min_BasePay='min',
    Average_BasePay='mean',
    Max_BasePay='max'
).sort_values(by='Average_BasePay', ascending=False)
```

```python
print("Top 10 Job Titles by Average BasePay:")
print(basepay_summary.head(10).to_markdown(numalign="left", stralign="left"))
```

```python
print("\n" + "="*80)
print("2. Averaging Numeric Values for Duplicated Employee Names")
print("="*80)
```

```python
df_no_id = df.drop(columns=['Id'], errors='ignore')
```

```python
# Identify duplicated employee names, excluding 'Not provided'
duplicated_names = df[df.duplicated(subset=['EmployeeName'], keep=False) & (df['EmployeeName'] != 'Not provided')]

# Check if there are any duplicated names to process
if not duplicated_names.empty:
    # Select only numeric columns for averaging
    numeric_cols = duplicated_names.select_dtypes(include=np.number).columns.tolist()

    # Ensure 'EmployeeName' is not in numeric_cols before grouping and averaging
    if 'EmployeeName' in numeric_cols:
        numeric_cols.remove('EmployeeName')

    # Average the numeric values for duplicated employee names
    averaged_duplicates = duplicated_names.groupby('EmployeeName')[numeric_cols].mean().reset_index()

    # Create a new simple ID column (1, 2, 3...)
    averaged_duplicates.insert(0, 'New_ID', range(1, 1 + len(averaged_duplicates)))

    print("Averaged numeric values for duplicated employee names:")
    display(averaged_duplicates.head().to_markdown(index=False, numalign="left", stralign="left"))

    print(f"Found {len(duplicated_names['EmployeeName'].unique())} unique employee names with duplicate entries.")
    print("Example of the new aggregated data (first 5 rows):")
    # Display the first 5 aggregated records
    print(averaged_duplicates.head(5).to_markdown(index=False, numalign="left", stralign="left"))
else:
    print("No duplicated employee names found (excluding 'Not provided').")
```

```python
print("\n" + "="*80)
print("3. Overtime Analysis")
print("="*80)

# Employees who did overtime are those with OvertimePay > 0
overtime_emps = df[df['OvertimePay'] > 0]

count_overtime_emps = len(overtime_emps)
total_overtime_pay = overtime_emps['OvertimePay'].sum()

print(f"Number of employees who did overtime: {count_overtime_emps}")
print(f"Total amount paid for overtime: \${total_overtime_pay:,.2f}")

# Simple Analysis: Top 5 Job Titles by Total Overtime Pay
top_overtime_jobs = overtime_emps.groupby('JobTitle')['OvertimePay'].sum().nlargest(5)
print("\nTop 5 Job Titles by Total Overtime Pay:")
print(top_overtime_jobs.to_markdown(numalign="left", stralign="left"))

print("\n--- What analysis can be done? ---")
print("A deeper analysis could involve:")
print("1. **Correlation:** Check the correlation between `OvertimePay` and `TotalPay` or `BasePay`.")
print("2. **Distribution:** Analyze the distribution of `OvertimePay` across different `Year`s to see if overtime hours changed over time.")
print("3. **Top Earners:** Identify the job titles or individuals receiving the highest amount of overtime pay to understand who is working the most extended hours.")
```

```python
# Calculate the average BasePay for the earliest and latest years
average_basepay_earliest = df[df['Year'] == df['Year'].min()]['BasePay'].mean()
average_basepay_latest = df[df['Year'] == df['Year'].max()]['BasePay'].mean()

# Calculate the overall percentage change
if average_basepay_earliest != 0:
    overall_change_pct = ((average_basepay_latest - average_basepay_earliest) / average_basepay_earliest) * 100
else:
    overall_change_pct = 0 # Handle the case where the earliest average is zero

if overall_change_pct > 0:
    print(f"\nConclusion: Yes, the average BasePay **increased** from the earliest year to the latest year.")
else:
    print(f"\nConclusion: No, the average BasePay **decreased** or remained flat from the earliest year to the latest year.")

print(f"The overall percentage change in average BasePay is: {overall_change_pct:.2f}%")
```
